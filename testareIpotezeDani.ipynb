```python
import pyphen
syllaber = pyphen.Pyphen(lang='ro')

def count_syllables(word):
    hyphenated = syllaber.inserted(word)
    return len(hyphenated.split('-'))
    
df['syllable_count'] = df['token'].apply(count_syllables)

corr, p_value = pearsonr(df['syllable_count'], df['score'])
print(f"Coeficient Pearson: {corr:.4f}, p-valoare: {p_value:.4g}")
```

```python
num_dependents_list = []
distance_to_root_list = []
depth_list = []

# Iterează prin rândurile din DataFrame
for _, row in df.iterrows():
    sentence = row['sentence']
    target_token = row['token'].lower()

    doc = nlp(sentence)

    found = False
    for token in doc:
        if token.text.lower() == target_token:
            # 1. Număr de copii (dependenți)
            num_dependents = len(list(token.children))

            # 2. Distanță până la root
            distance_to_root = abs(token.i - token.head.i)

            # 3. Adâncime în arbore
            depth = 0
            parent = token
            while parent != parent.head:       
                depth += 1
                parent = parent.head

            # Adaugă în liste
            num_dependents_list.append(num_dependents)
            distance_to_root_list.append(distance_to_root)
            depth_list.append(depth)
            found = True
            break

    # Dacă nu găsește tokenul (fallback)
    if not found:
        num_dependents_list.append(None)
        distance_to_root_list.append(None)
        depth_list.append(None)

print(num_dependents_list) 
corr, p_value = pearsonr(num_dependents_list, df['score'])
print(f"Coeficient Pearson: {corr:.4f}, p-valoare: {p_value:.4g}")
print(distance_to_root_list)
corr, p_value = pearsonr(distance_to_root_list, df['score'])
print(f"Coeficient Pearson: {corr:.4f}, p-valoare: {p_value:.4g}")
print(depth_list) 
corr, p_value = pearsonr(depth_list, df['score'])
print(f"Coeficient Pearson: {corr:.4f}, p-valoare: {p_value:.4g}")
```

```python
class KNNPredictor:
    def __init__(self, knowledge_df, k=3):
        self.k = k
        self.base = [
            (row["token"], row["score"], nlp(row["token"]).vector)
            for _, row in knowledge_df.iterrows()
        ]

    def predict(self, word):
        v = nlp(word).vector
        distante = [(token, score, np.sum((v - vec) ** 2)) for token, score, vec in self.base]
        top_k = sorted(distante, key=lambda x: x[2])[:self.k]
        return sum(score for _, score, _ in top_k) / self.k

# 📄 împărțim datele în train/test
train_df, test_df = train_test_split(df[["token", "score"]], test_size=0.3, random_state=42)

# 🧠 antrenăm KNN pe train
knn = KNNPredictor(train_df, k=200)

# 🔍 testăm pe test
y_true = []
y_pred = []

for _, row in test_df.iterrows():
    token = row["token"]
    scor_real = row["score"]
    scor_pred = knn.predict(token)
    y_true.append(scor_real)
    y_pred.append(scor_pred)
    print(f"{token}: real={scor_real:.2f}, pred={scor_pred:.2f}")

# 📊 evaluare performanță
r2 = r2_score(y_true, y_pred)
mae = mean_absolute_error(y_true, y_pred)

print(f"\nR²: {r2:.4f}")
print(f"MAE: {mae:.4f}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import pearsonr, spearmanr
import numpy as np

import warnings
warnings.filterwarnings("ignore", message="An input array is constant; the correlation coefficient is not defined.")


def pearson_r2(preds, y_true):
    """
    Custom lexical complexity prediction metric consisting in:
    - `(max(0, R^2) + abs(pearson))/2`
    - R^2 - is the standard regression metric, here we do not allow negative values
    - pearson - is the pearson correlation

    A good sollution would:
    - have values that are close to the actual lexical complexity scores
    - have values that are correlated well to the complexity scores
    - one can obtain a good correlation with values that are outside of the range or inversely proportional, we don't care as much
    - a constant predictor will get 0, pearson will be NaN

    """
    r2 = r2_score(y_true, preds, sample_weight=None, force_finite=True)
    r2 = max(0, r2)
    pears = pearsonr(y_true, preds)[0]
    pears = np.nan_to_num(np.array([pears]), copy=False, nan=0)[0]
    pears = np.abs(pears)
    return (pears + r2)/2

def evaluate(predictions, y_true):
    cust = pearson_r2(predictions, y_true)
    return {'r2_pearson': cust}
```
